```python
df.groupby('category')[['review_rating']].median()
```

```python
df['review_rating']=df.groupby('category')[['review_rating']].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#create a new column 'age_group':

df['age_group']=pd.qcut(df['age'],q=4,labels=['Young Adult','Adult','Middle-aged','Senior'])
```

```python
#create a new column called 'purchase_frequency_days':

mapping_dict={ 'Fortnightly':14,
               'Weekly':7,
               'Monthly':30,
               'Quarterly':90,
               'Bi-Weekly':14,
               'Annually':365,
               'Every 3 months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(mapping_dict)
```
